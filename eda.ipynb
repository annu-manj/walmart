Replace the null value with category based median

```python
df[df['unit_price'].isnull()]
```

```python
# First, clean the unit_price column by removing '$' and converting to float
df['unit_price'] = df['unit_price'].str.replace('$', '', regex=False).astype(float)

# Then fill null values with category-based median
df['unit_price'] = df.groupby('category')['unit_price'].transform(lambda x: x.fillna(x.median()))
```

```python
df['quantity'].unique()
```

```python
df['quantity'] = df.groupby('category')['quantity'].transform(lambda x:x.fillna(x.median()))
```
